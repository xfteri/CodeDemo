```python
# 需求：
# 加载 data/billboard.csv，将歌曲周排行数据集拆分成2个数据集：
# 数据集1：保存歌曲的基本信息
# 数据集2：保存歌曲的每周排行信息
# 要求两个数据集以一个 id 列，在逻辑上互相关联
# 加载并观察数据集
```

```python
bill_board = pd.read_csv('./data/billboard.csv')
bill_board.head()
```

实现思路如下：
- 变换为长数据集
- 提取歌曲信息表并添加歌曲id列
- 长数据集添加歌曲id
- 提取每周评分数据表

```python
# 变换为长数据集
billboard_long =pd.melt(bill_board,
                  id_vars=['year','artist','track','time','date.entered'],
                  var_name='week',
                  value_name='rating')
billboard_long
```

```python
# 取出指定的列，生成歌曲信息表
billboard_songs = billboard_long[['year','artist','track','time','date.entered']]
billboard_songs
```

```python
# 以歌曲名字Loser为例，发现歌曲信息表中有很多的重复值
billboard_songs[billboard_songs.track=='Loser']
```

```python
# 歌曲信息去重
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs
```

```python
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs
```

## 长数据集添加歌曲id

### 将歌曲信息表中的歌曲id关联到原始的长数据集中


```python
# 原长数据集与歌曲信息表进行merge合并，基于指定的字段，这样就歌曲id放到了长数据集中
new_billboard_long = billboard_long.merge(billboard_songs,on = ['year','artist','track','time','date.entered'])
new_billboard_long
```

## 提取每周评分数据表

### 从有歌曲id的长数据集中，提取每周评分数据表

```python
billboard_ratings = new_billboard_long[['id','week','rating']]
billboard_ratings
```

## 拆分数据集的作用

```python
billboard_songs.merge(billboard_ratings,on=['id'])
```

```python
# 好处：拆分后的 2 个数据集在内存占用上，也比长数据集要少的多
print(billboard_songs.info(memory_usage='deep'))
print('='*40)
print(billboard_ratings.info(memory_usage='deep'))
print('='*40)
print(bill_board.info(memory_usage='deep'))
print('='*40)
print(billboard_long.info(memory_usage='deep'))
```

# stack 整理数据

## stack 和 unstack 简介

## stack 功能演示

```python
# 加载 state_fruit.csv 数据集
state_fruit = pd.read_csv('./data/state_fruit.csv',index_col=0)
state_fruit
```

```python
# 使用 stack 函数整理数据，查看效果
state_fruit_series = state_fruit.stack()
state_fruit_series
```

```python
# 使用 reset_index()，将结果变为 DataFrame
state_fruit_tidy = state_fruit_series.reset_index()
state_fruit_tidy
```

```python
# 给 columns 重新命名
state_fruit_tidy.columns = ['state','fruit','weight']
state_fruit_tidy
```

```python
# 也可以使用 rename_axis 给不同的行标签索引层级命名
state_fruit.stack().rename_axis(['state','fruit'])
```

```python
# 再次使用 reset_index
state_fruit.stack().rename_axis(['state','fruit']).reset_index()
```

```python
state_fruit.stack().rename_axis(['state','fruit']).reset_index(name='weight')
```

## 使用 unstack 函数

```python
state_fruit_series.unstack()
```

# wide_to_long 整理数据

需求：

- 加载data/movie.csv数据，统计每部电影的每个主演的被点赞数，返回新的df
- 新df 的列名为 movie_title、actor_num、actor、actor_facebook_likes：
- 分别代表电影名称、演员编号、演员姓名、该演员被点赞数俊辉

## 初步整理数据

```python
# 加载movie.csv数据
movies = pd.read_csv('./data/movie.csv')
movies.head()
```

## 去除无关字段


```python
movie_actors = movies[['movie_title','actor_1_name','actor_2_name','actor_3_name','actor_1_facebook_likes','actor_2_facebook_likes','actor_3_facebook_likes']]
movie_actors.head()
```

## 整理colmns列名

```python
movie_actors.columns = ['movie_title','actor_name_1','actor_name_2','actor_name_3',
                       'actor_facebook_likes_1','actor_facebook_likes_2','actor_facebook_likes_3']
movie_actors.head()
```

# wide_to_long函数的具体使用

```python
movie_actors_tidy = pd.wide_to_long(movie_actors,
                                   stubnames=['actor_name','actor_facebook_likes'],
                                   i = ['movie_title'],
                                   j = 'actor_num',
                                   sep='_').reset_index()
movie_actors_tidy.head()
```

```python
movie_actors_tidy[movie_actors_tidy.movie_title=='Avatar']
```

# 总结：
- melt、stack、wide_to_long函数均可以实现将宽数据整理成长数据
- melt：指定数据列，将指定列变成长数据
- stack：返回一个具有多层级索引的数据，配合reset_index可以实现宽数据变成长数据
- wide_to_long：处理列名带数字后缀的宽数据
- stack/unstack, melt/pivot_table 互为逆向操作-
