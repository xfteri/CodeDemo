```python
import pandas as pd
```

# 1 DataFrame行操作

## 1.1 新增行/修改行

```python
scientists = pd.read_csv('./data/scientists.csv')
scientists
```

```python
# 准备新行数据
new_series = pd.Series(['LuoGeng Hua', '1910-11-12', '1985-06-12', 75, 'Mathematician'],
                       index=['Name', 'Born', 'Died', 'Age', 'Occupation'])
scientists.append(new_series,ignore_index=True)
```

## 1.1修改行

### 修改行时，时直接对原始DataFrame进行修改

```python
scientists.loc[4] = ['Rachel Carson','1907-05-27','1964/04/14',56,'Biologist']
scientists
```

```python
# 修改行标签为4的行的Born和Age列的数据
scientists.loc[4,['Born','Age']] = ['1906-5-27',58]
scientists
```

```python
# 修改行标签为6的行的Born列的数据为1912-07-23
scientists.loc[6,['Born']] = '1912-07-23'
scientists
```

## 1.2删除行

### 删除行时，会返回新的DataFram

```python
# 删除行标签为1和3的行
scientists.drop([1,3])
```

# 2 DataFrame列操作

## 2.1新增列/修改列

```python
# 给scientists数据增加一个Country列
scientists['Country'] = ['England', 'England', 'England', 'French', 
                         'America', 'England', 'England', 'Germany']
scientists
```

```python
scientists.loc[:,'Country'] = ['England', 'England', 'England', 'French', 
                         'America', 'England', 'England', 'Germany']
scientists
```

```python
# 修改 scientists 数据中 Country 列的数据
scientists['Country'] = ['england', 'england', 'england', 'french', 
                         'america', 'england', 'england', 'germany']
scientists
```

```python
scientists.loc[:, 'Country'] = ['england', 'england', 'england', 'french', 
                                'america', 'england', 'england', 'germany']
scientists
```

## 2.2删除列

### 删除列时，会返回新的DataFrame

```python
# 删除scientists数据中Country列的数据
scientists.drop(['Country'],axis=1)
```
